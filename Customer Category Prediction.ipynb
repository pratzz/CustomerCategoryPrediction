```python
train = train.dropna()
train.shape
```

```python
from sklearn.model_selection import train_test_split

# Splitting data into train and validation sets
train_split, valid = train_test_split(train, train_size = 0.8)
train_y =  train_split["customer_category"]
train_x = train_split.loc[:, train_split.columns != 'customer_category']
valid_y =  valid["customer_category"]
valid_x = valid.loc[:, valid.columns != 'customer_category']
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(train_x)
scaled_train_x = scaler.transform(train_x)
scaled_valid_x = scaler.transform(valid_x)
```

```python
from sklearn.linear_model import LogisticRegression 
classifier = LogisticRegression(random_state = 0) 
classifier.fit(scaled_train_x, train_y) 
```

```python
from sklearn.metrics import precision_score

train_pred = classifier.predict(scaled_train_x) 
valid_pred = classifier.predict(scaled_valid_x) 

train_score = precision_score(train_y, train_pred, average='macro')
valid_score = precision_score(valid_y, valid_pred, average='macro')
print("Train score:", train_score)
print("Validation score:", valid_score)
```

```python
from sklearn import linear_model, decomposition, datasets
from sklearn.pipeline import Pipeline
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
    
std_slc = StandardScaler()
pca = decomposition.PCA()
logistic_Reg = linear_model.LogisticRegression(solver='liblinear')

pipe = Pipeline(steps=[('std_slc', std_slc),
                           ('pca', pca),
                           ('logistic_Reg', logistic_Reg)])

n_components = list(range(1,train_x.shape[1]+1,1))
C = np.logspace(-4, 4, 50)
penalty = ['l1', 'l2']
parameters = dict(pca__n_components=n_components,
                      logistic_Reg__C=C,
                      logistic_Reg__penalty=penalty)

clf_GS = GridSearchCV(pipe, parameters, n_jobs=-1)
clf_GS.fit(train_x, train_y)
    
print('Best Penalty:', clf_GS.best_estimator_.get_params()['logistic_Reg__penalty'])
print('Best C:', clf_GS.best_estimator_.get_params()['logistic_Reg__C'])
print('Best Number Of Components:', clf_GS.best_estimator_.get_params()['pca__n_components'])
print(); print(clf_GS.best_estimator_.get_params()['logistic_Reg'])
```

```python
clf_GS.best_estimator_
```

```python
from sklearn.metrics import precision_score

train_pred = clf_GS.best_estimator_.predict(train_x) 
valid_pred = clf_GS.best_estimator_.predict(valid_x) 

train_score = precision_score(train_y, train_pred, average='macro')
valid_score = precision_score(valid_y, valid_pred, average='macro')
print("Train score:", train_score)
print("Validation score:", valid_score)
```

```python
from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(train_y, train_pred)) 
print ("Accuracy : ", accuracy_score(valid_y, valid_pred)) 
```
